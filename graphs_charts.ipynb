```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ww_df = pd.read_csv(r"C:\Users\israe\Desktop\Files\Coding\Repos\WW_Project\ww_boxes_new.csv", index_col = 0)
ww_df.tail()
```

```python
west_side = ww_df[ww_df['Address'].str.contains(r'\bNW\b|\bW\b|\bSW\b', case=False, regex=True)]

# Print the first 10 rows of the filtered DataFrame
west_side.head(10)
```

```python
east_side = ww_df[ww_df['Address'].str.contains(r'\bNE\b|\bE\b|\bSE\b', case=False, regex=True)]

# Print the first 10 rows of the filtered DataFrame
east_side.tail(10)
```

```python

# Counting the number of east side addresses
east_side_count = ww_df[ww_df['Address'].str.contains(r'\bNE\b|\bE\b|\bSE\b', case=False, regex=True)].shape[0]

# Counting the number of west side addresses
west_side_count = ww_df[ww_df['Address'].str.contains(r'\bNW\b|\bW\b|\bSW\b', case=False, regex=True)].shape[0]

# Define the side labels
side_labels = ['East Side', 'West Side']

# Define the address counts
address_counts = [east_side_count, west_side_count]

# Creating a bar chart
plt.bar(side_labels, address_counts, color=['blue', 'green'])

# Adding labels and title
plt.xlabel('Side')
plt.ylabel('Number of Boxes')
plt.title('Comparison of East Side vs. West Side')

# Displaying the plot
plt.show()

```

```python
# Counting the number of east side addresses
north_side_count = ww_df[ww_df['Address'].str.contains(r'\bNE\b|\bNW\b|\bN\b', case=False, regex=True)].shape[0]

# Counting the number of west side addresses
south_side_count = ww_df[ww_df['Address'].str.contains(r'\bSW\b|\bS\b|\bSE\b', case=False, regex=True)].shape[0]

# Define the side labels
side_labels = ['North Side', 'South Side']

# Define the address counts
address_counts = [north_side_count, south_side_count]

# Creating a bar chart
plt.bar(side_labels, address_counts, color=['blue', 'green'])

# Adding labels and title
plt.xlabel('Side')
plt.ylabel('Number of Boxes')
plt.title('Comparison of North Side vs. South Side')

# Displaying the plot
plt.show()
```
